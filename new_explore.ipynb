```python
import numpy as np
import pandas as pd
```

```python
df_orders = pd.read_csv("ecommerce_data//olist_orders_dataset.csv")
df_items = pd.read_csv("ecommerce_data//olist_order_items_dataset.csv")
df_products = pd.read_csv("ecommerce_data//olist_products_dataset.csv")
```

```python
df_customers = pd.read_csv("ecommerce_data//olist_customers_dataset.csv")
df_customers.head()
```

```python
df_merge_1 = df_items.merge(df_products, on='product_id', how='left')
df_merge_1
```

## Translate Product Category

```python
df_product_translate = pd.read_csv("ecommerce_data//product_category_name_translation.csv")
df_product_translate
```

```python
trans_dict = dict(zip(df_product_translate.product_category_name, df_product_translate.product_category_name_english))
len(trans_dict)
```

```python
df_merge_1['product_category_name_eng'] = df_merge_1['product_category_name'].apply(lambda x: trans_dict.get(x,0))
df_merge_1
```

```python
df_merge_2 = df_merge_1.merge(df_orders, on='order_id', how='left')
df_merge_2
```

```python
df_merge_3 = df_merge_2.merge(df_customers, on='customer_id', how='left')
df_merge_3
```

```python
df_merge_3['order_status'].value_counts()
```

```python
df_merge_3 = df_merge_3[(df_merge_3['order_status'] != 'unavailable') & (df_merge_3['order_status'] != 'canceled')]
df_merge_3
```

```python
df_merge_3['order_status'].value_counts()
```

```python
df_merge_3['order_purchase_timestamp'] = pd.to_datetime(df_merge_3['order_purchase_timestamp'])
df_merge_3.info()
```
